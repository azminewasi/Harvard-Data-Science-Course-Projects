Random Graph

```python

def er_graph (N,p):
    G = nx.Graph()
    G.add_nodes_from(range(N))

    for node1 in G.nodes():
        for node2 in G.nodes():
            if node1<node2 and bernoulli.rvs(p=p):
                G.add_edge(node1, node2)
    return G
```

```python
nx.draw(er_graph(50,0.08),node_color="blue",node_size=40,edge_color="red")
plt.savefig("ER-1.pdf")
```
